```python
#importing basics libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# import the traing and test data and convert to dataframes
train_df = pd.read_excel(r"D:\ds_resources\flight_data\Data_Train.xlsx")
test_df = pd.read_excel(r"D:\ds_resources\flight_data\Test_set.xlsx")
```

```python
# Append the test data to the train data to create a new df
df= train_df.append(test_df)
```

```python
# get the basic information about the data
df.info()
```

```python
# All the objects have to be converted to int or floats for the models to understand
```

```python
"""
Start with 'Date_of_Journey'. Here the data of journey is given as an object, we will convert it into day, month and year.
After that append the day, month and year of journey columns to the df and delete the original column.
"""
df['Date_of_Journey']
```

```python
# We have to split on the forward slash.
df['Date_of_Journey'].str.split('/')
```

```python
# Convert the above to a string type and select 0th index for date, 1st for month and 2nd for year.
df['Date_journey']=df['Date_of_Journey'].str.split('/').str[0]
df['Month_journey']=df['Date_of_Journey'].str.split('/').str[1]
df['Year_journey']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
# Now we can drop the 'Date_of_Journey' column
df.drop(['Date_of_Journey'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# convert date, month and year of journey into int
df['Date_journey']=df['Date_journey'].astype(int)
df['Month_journey']=df['Month_journey'].astype(int)
df['Year_journey']=df['Year_journey'].astype(int)
```

```python
df.info()
```

```python
# Convert arrival time to arrival hour and minute
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
# drop arrival time
df.drop(['Arrival_Time'], axis=1, inplace = True)
```

```python
df.head()
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.head()
```

```python
# do the same operations for departure time
df['Dept_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dept_min']=df['Dep_Time'].str.split(':').str[1]
df['Dept_hour']=df['Dept_hour'].astype(int)
df['Dept_min']=df['Dept_min'].astype(int)
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
# For total stops check the unique values and map values to it
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.drop(9039, inplace=True, axis = 0)
```

```python
df
```

```python
df.info()
```

```python
df['duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.drop(['Duration'], axis =1 , inplace = True)
```

```python
df['duration_hour']=df['duration_hour'].astype(int)
df['duration_min']=df['duration_min'].astype(float)
```
