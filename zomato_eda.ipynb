```python
country_df = pd.read_excel(r"D:\ds_resources\Country-Code.xlsx")
```

```python
country_df.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, country_df, on='Country Code', how='left')
```

```python
final_df
```

```python
# to check the different data types 
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts().index
country_values = final_df.Country.value_counts().values
```

```python
country_names, country_values
```

```python
## Pie Chart- Top 3 countries that uses zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
"""
Observation:
The max number of transactions of zomato are done in India, followed by the US and then the UK.
zomato is doing the best in India.
"""
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observations
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor
and so on

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15, 8)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
# to plot the graph with corresponsing colors for rating
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings, 
            palette=['blue','red','orange','yellow','green','green'])
```

Observation:

Not Rated count is very high

Maximum number of rating are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
#Find the country names where users have given 0 ratings
final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

```python
#Find out which currency is used by which counntry
final_df[['Country', 'Currency']].groupby(['Country', 'Currency']).size().reset_index()
```

```python
# Find which country has online deliver
final_df[['Country', 'Has Online delivery']].groupby(['Country', 'Has Online delivery']).size().reset_index()
```

```python
# Find the top 5 cities with most transactions
city_values = final_df.City.value_counts().values
city_names = final_df.City.value_counts().index
```

```python
city_values, city_names
```

```python
plt.pie(city_values[: 5], labels=city_names[: 5], autopct='%1.2f%%')
```
